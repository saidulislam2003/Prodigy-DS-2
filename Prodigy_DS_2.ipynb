```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/content/train.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.drop(columns = 'Cabin', axis = 1, inplace = True)
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
sns.countplot(x = 'Pclass', data = df)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
sns.countplot(x = 'Sex', data = df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize = (8, 8))
sns.histplot(df['Age'], bins = 20, kde = True)
plt.title("Age distribution")
plt.show()
```

```python
sns.barplot(x = 'Pclass', y = 'Survived', data = df)
plt.title("Passenger Class vs Survival Rate")
plt.show()
```

```python
sns.barplot(x = 'Sex', y = 'Survived', data = df)
plt.title("Gender VS Survival Rate")
plt.show()
```

```python
plt.figure(figsize = (8, 8))
sns.histplot(data = df,x = 'Age', hue = 'Survived', kde = True, multiple = 'stack')
plt.title("Age VS Survival Rate")
plt.show()
```

```python
sns.catplot(x = 'Pclass', y = 'Survived', hue = 'Sex', data = df, kind = 'bar')
plt.title("Passenger Class, Gender VS Survival Rate")
plt.show();
```

```python
plt.figure(figsize = (8, 6))
numeric = df.select_dtypes(include = [np.number])
correlation = numeric.corr()
sns.heatmap(correlation, annot = True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.barplot(x = 'SibSp', y = 'Survived', data = df)
plt.title("Siblings or Spouses Number VS Survival Rate")
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.barplot(x = 'Survived', y = 'Fare', data = df)
plt.title("Fare VS Survival Rate")
plt.show()
```

```python
sns.barplot(x = 'Embarked', y = 'Survived', data = df)
plt.xlabel("S = Southampton, C = Cherbourg, Q = Queenstown")
plt.title("Embarked VS Survival Rate")
plt.show()
```
